```python
import cv2
import mediapipe as mp
import numpy as np 
import time
```

```python
mp_face=mp.solutions.face_detection
face=mp_face.FaceDetection()
```

```python
width, height = 600, 400
blank_image = 255 * np.ones((height, width, 3), dtype=np.uint8)
font = cv2.FONT_HERSHEY_SIMPLEX
text = "Take Break For Your Safety"
font_scale = 1
font_thickness = 3
text_size = cv2.getTextSize(text, font, font_scale, font_thickness)[0]
text_x = (width - text_size[0]) // 2
text_y = (height + text_size[1]) // 2

```

```python
cap=cv2.VideoCapture(0)
max_time=20 #Dynamic
start_time =time.time()


while 1 :
    ret,frame=cap.read()
    if not ret :
        break 
    frame=cv2.cvtColor(frame,cv2.COLOR_BGR2RGB)
    out=face.process(frame)
    
    if out.detections : 
        c_time = int(time.time() - start_time)
        if c_time > max_time:
            cv2.putText(blank_image, text, (text_x, text_y), font, font_scale, (0, 0, 255), font_thickness, cv2.LINE_AA)
            cv2.imshow("Warning ",blank_image)
            
            

                
    else :
        start_time = time.time()
        
        print("no FACE ")
        cv2.destroyAllWindows()
        #cap.release()
    #frame=cv2.cvtColor(frame,cv2.COLOR_RGB2BGR)    
   # cv2.imshow("frame",frame)    
    key = cv2.waitKey(0)
    if key == ord("q"):
       
        break
cap.release()     
cv2.destroyAllWindows()            
        
        
        
```
